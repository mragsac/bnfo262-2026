## Interface

#### Kernel
- Restart the Kernel: Use this option if your notebook is stuck or you want to clear all variables and start fresh.

## Operations

#### Switching Modes:
- Press `Enter` to enter Edit Mode (where you can type inside a cell).
- Press `Esc` to exit Edit Mode and return to Command Mode (where you can use shortcuts to manipulate cells).

#### Running a Cell:
- Press `Shift-Enter` to execute the code or render text in the current cell and automatically move to the next cell.

#### Changing Cell Type:
- Press `y` to change the current cell to a Code cell (used for writing code).
- Press `m` to change the current cell to a Markdown cell (used for writing formatted text, headings, lists, etc.).

#### Adding New Cells:
- Press `a` to create a new cell above the current one.
- Press `b` to create a new cell below the current one.

#### Deleting a Cell:
- Press d twice (`d-d`) while in Command Mode to delete the current cell.

# Comparison Operators

Comparison operators produce a boolean result

i.e. 4 > 3 -> True whereas 4 < 3 -> False

```python
result = 4 > 3
print(result)
print(type(result))
print()

# Equal
result = 4 == 3
print("4==3:", result)

# Not Equal
result = 4 != 3
print("4!=3: " + str(result))

# Greater than
result = 4 > 3
print(f"4>3: {result}")

# Less than
print(f"4<3: {4 < 3}")

# Greater than or equal to
result = 4 >= 3
print(f"4>=3: {result}")

# Less than or equal to
result = 4 <= 3
print(f"4<=3: {result}")
```

## Conditional Statements

Conditional statements allow your program to make decisions based on certain conditions.

- `if`: Executes a block of code if the condition is `True`.
- `elif`: (short for "else if") Checks another condition if the previous `if` or `elif` was `False`.
- `else`: Executes a block of code if all previous conditions were `False`.

```python
today = "friday"

if today == "friday":
    print("Thank goodness it's Friday!")
elif today == "thursday":
    print("One more day to Friday!")
else:
    print("Can't wait until it's Friday!")
```

## Looping and Iteration

Looping allows you to execute a block of code repeatedly, either a specific number of times or until a condition is met.

- `for` Loop: Used for iterating over a sequence (e.g., list, tuple, string, range, dictionary). Executes a block of code for each element in the sequence.

- `while` Loop: Repeats as long as a specified condition evaluates to `True`. Useful when the number of iterations is not predetermined.

```python
# for loop

s = [1, 2, 3, 4]
for i in s:
    print(i)
```

```python
# while loop

k = 0
while k < 10:
    print(k)
    k += 1
```

# List Comprehensions (Advanced)

List comprehensions are a frequently used shorthand way to create lists which involve iterating through another object

Format:  
- simple: <code>[{do something with x} for x in {iterable_object}]</code>
- filtering: <code>[x for x in {iterable_object} if {logical test on x}]</code>
  - conditional expression after <code>for</code> clause
- modifying x based on criteria: <code>[{value if True} if {logical test on x} else {value if False} for x in {iterable_object}]</code>
  - conditional expression before <code>for</code> clause

Fun tip: This can be used for other data structures (i.e. set comprehension, dictionary comprehensions)

```python
# create a list using a for loop
l = []
for i in range(10):
    l.append(i)
print(l)

# same result with list comprehension
l = [x for x in range(10)]
print(l)
```

```python
# SIMPLE: create a sequence x^2 from 0 to 9
l = [x**2 for x in range(10)]
print(l)

# FILTER: create list with only numbers above 5
# conditional expression AFTER for clause
l = [x for x in range(10) if x > 5]
print(l)

# MODIFY: change negative values to 0 
# conditional expression BEFORE for clause
l = [0 if x < 0 else x for x in range(-3,4)]
print(l)
```
